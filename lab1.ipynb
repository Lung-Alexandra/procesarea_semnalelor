```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

pdf_pages = PdfPages('lab1.pdf')
```

```python
# ex1 a,b
t = np.arange(0, 0.03, 0.0005)  # Axa timpului de la 0 la 0.03 secunde
nr_subplots = 3
# Definirea semnalelor continue
x_t = np.cos(520 * np.pi * t + np.pi / 3)
y_t = np.cos(280 * np.pi * t - np.pi / 3)
z_t = np.cos(120 * np.pi * t + np.pi / 3)
fig, axs = plt.subplots(nr_subplots)
fig.suptitle('1.b) Semnale continue x(t), y(t) si z(t)')
axs[0].plot(t, x_t)
axs[1].plot(t, y_t)
axs[2].plot(t, z_t)
pdf_pages.savefig()
plt.show()
```

```python
t = np.linspace(0, 1, 200)
nr_subplots = 3

# Definirea semnalelor continue
x_t = np.cos(520 * np.pi * t + np.pi / 3)
y_t = np.cos(280 * np.pi * t - np.pi / 3)
z_t = np.cos(120 * np.pi * t + np.pi / 3)
fig, axs = plt.subplots(nr_subplots)
fig.suptitle('1.c) Semnale  x[t], y[t] si z[t]')
axs[0].set_xlim([0, 0.03])
axs[1].set_xlim([0, 0.03])
axs[2].set_xlim([0, 0.03])
t_c = np.arange(0,0.03, 0.0005)
x_c = np.cos(520 * np.pi * t_c + np.pi / 3)
y_c = np.cos(280 * np.pi * t_c - np.pi / 3)
z_c = np.cos(120 * np.pi * t_c + np.pi / 3)

axs[0].plot(t_c, x_c)
axs[0].stem(t, x_t)
axs[1].plot(t_c, y_c)
axs[1].stem(t, y_t)
axs[2].plot(t_c, z_c)
axs[2].stem(t, z_t)
pdf_pages.savefig()

plt.show()
```

```python
# ex 2
# (a) Semnal sinusoidal de frecvență 400 Hz cu 1600 de eșantioane

t = np.linspace(0,1,1600)
signal = np.sin(2*np.pi*400*t)
t_c = np.linspace(0,0.03,1000)
signal_c = np.sin(2*np.pi*400*t_c)
fig, axs = plt.subplots(1)
fig.suptitle('2.a) Semnal sinusoidal de 400 Hz cu 1600 de eșantioane')
axs.set_xlim([0, 0.03])
axs.stem(t, signal)
axs.plot(t_c,signal_c)
pdf_pages.savefig()

plt.show()
```

```python

# (b) Semnal sinusoidal de frecvență 800 Hz pentru 3 secunde

t = np.linspace(0,3,100)
signal = np.sin(2*np.pi*800*t)
fig, axs = plt.subplots(2)
fig.suptitle('2.b) Semnal sinusoidal de 800 Hz pentru 3 secunde')
axs[0].set_title("Numar esantioane = 10^2")
axs[0].plot(t, signal)
t1 = np.linspace(0,3,1000)
signal1 = np.sin(2*np.pi*800*t1)
axs[1].set_title('Numar esantioane = 10^3')
axs[1].plot(t1, signal1)
plt.tight_layout()
pdf_pages.savefig()


plt.show()

```

```python

# (c) Semnal de tip sawtooth de frecvență 240 Hz
# p = 1/frecventa
# 2(t*frecv-[0.5+t*frecv])

t= np.linspace(0,0.03,1000)
signal = 2*(t*240-np.floor(0.5+t*240))
fig, axs = plt.subplots(1)
fig.suptitle('2.c) Semnal de tip sawtooth de 240 Hz')
axs.plot(t, signal)
pdf_pages.savefig()

plt.show()

# # (d) Semnal de tip square de frecvență 300 Hz
t= np.linspace(0,0.03,1000)
signal = np.sign(np.sin(2*np.pi*300*t))
# signal = 2*(2*np.floor(300*t)-np.floor(2*300*t))+1
fig, axs = plt.subplots(1)
fig.suptitle('2.d) Semnal de tip square de frecvență 300 Hz')
axs.plot(t, signal)
pdf_pages.savefig()

plt.show()
```

```python

# (e) Semnal 2D aleator de dimensiune 128x128
x, y = 128, 128
signal = np.random.rand(x, y)
plt.imshow(signal, cmap='gray')
plt.colorbar()
plt.title('2.e) Semnal 2D aleator de dimensiune 128x128')
pdf_pages.savefig()

plt.show()

# (f) Semnal 2D personalizat de dimensiune 128x128

x, y = 128, 128
signal = np.zeros((x, y))
for i in range(x):
    for j in range(y):
        signal[i, j] = i / x #Valoarea semnalului crește pe măsură ce trecem pe verticală
plt.imshow(signal, cmap='gray')
plt.colorbar()
plt.title('2.f) Semnal 2D personalizat de dimensiune 128x128')
pdf_pages.savefig()

plt.show()
pdf_pages.close()
```
